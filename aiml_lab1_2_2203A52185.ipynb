```python
import pandas as pd
data = {
    "Name": ["kiran", "arun", "vijay", "varun"],
    "Age": [25, 30, 35, 40],
    "Height": [145, 151, 165, 173],
    "Weight": [45, 55, 65, 75]
}
df = pd.DataFrame(data)
print(df)
```

```python
import matplotlib.pyplot as plt
# 3. Scatter Plot
plt.figure(figsize=(8, 6))
plt.scatter(df["Age"], df["Height"], c="blue", alpha=0.7)
plt.title("Scatter Plot: Age vs. Height")
plt.xlabel("Age (Years)")
plt.ylabel("Height (cm)")
plt.grid(True)
plt.tight_layout()
plt.show()
# 4. Bar Chart
plt.figure(figsize=(8, 6))
plt.bar(df["Name"], df["Weight"], color="skyblue")
plt.title("Bar Chart: Weight by Name")
plt.xlabel("Name")
plt.ylabel("Weight (kg)")
plt.xticks(rotation=45, ha="right")
plt.grid(axis="y")
plt.tight_layout()
plt.show()
# 5. Histogram
plt.figure(figsize=(8, 6))
plt.hist(df["Age"], bins=10, edgecolor="black")
plt.title("Histogram: Age Distribution")
plt.xlabel("Age (Years)")
plt.ylabel("Frequency")
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```
